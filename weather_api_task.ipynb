## Define the Function

```python
def get_city_data(city, api_key):
    baseurl = "http://api.weatherapi.com/v1/"
    
    #Get current weather
    weather_url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}"
    weather_response = requests.get(weather_url)
    if weather_response.status_code == 200:
        data = weather_response.json()
        current = data['current']
        location = data['location']
        temp_f = current['temp_f']
        wind_mph = current['wind_mph']
        humidity = current['humidity']
        
    else:
        print(f"Error fetching current weather data. There is no weather for {city}: {response.status_code}")

    # Get forecast
    forecast_url = f"{baseurl}forecast.json?key={api_key}&q={city}&days=3"
    forecast_response = requests.get(forecast_url)
    
    if forecast_response.status_code == 200:
        forecast_data = forecast_response.json()
        forecast = forecast_data['forecast']['forecastday']
        
        forecast_3_day_summary = []
        for day in forecast:
            date_obj = datetime.strptime(day['date'], '%Y-%m-%d')
            formatted_date = date_obj.strftime('%A,%B %d')
            forecast_3_day_summary.append(f"{formatted_date}: High {day['day']['maxtemp_f']}F / Low: {day['day']['mintemp_f']}F - {day['day']['condition']['text']}")
    else:
        forecast_3_day_summary = "No forecast available!"
 
    # Get astronomy data
    today = pd.Timestamp.now().strftime('%Y-%m-%d')
    astronomy_url = f"{baseurl}astronomy.json?key={api_key}&q={city}&dt={today}"
    astronomy_response = requests.get(astronomy_url)

    if astronomy_response.status_code == 200:
        astro_data = astronomy_response.json()
        astro = astro_data['astronomy']['astro']
        
        astro_sunrise = astro['sunrise']
        astro_sunset = astro['sunset']
        
    else:
        astro_sunrise = "No sunrise data at this time!"
        astro_sunset = "No sunset data at this time!"
 
    # Get alerts data
    alerts_url = f"{baseurl}alerts.json?key={api_key}&q={city}"
    alerts_response = requests.get(alerts_url)
 
    if alerts_response.status_code == 200:
        alerts_data = alerts_response.json()
        alerts = alerts_data['alerts']['alert']
        
        alerts_info = []
        for alert in alerts:
            alerts_info.append(f"Alert: {alert['headline']}, Certainty: {alert['certainty']}, Description: {alert['desc']}")
    
    else:
        alerts_info = "No weather alerts at the moment. You're safe!"
 
    
    # Build a summary dictionary of city/weather data
    city_summary = {
        "City": f"{location['name']}, {location['region']}",
        "Temperature": f"{current['temp_f']}F / {current['temp_c']}C",
        "Condition": current['condition']['text'],
        "Humidity": f"{current['humidity']}%",
        "Wind Speed": f"{current['wind_mph']} mph",
        "Forecast": forecast_3_day_summary,
        "Sunrise": astro_sunrise,
        "Sunset": astro_sunset,
        "Alerts": alerts_info
    }

    return city_summary

   
```
